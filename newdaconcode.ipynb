```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(method='bfill')
```

```python
df.isnull().sum()
```

```python
df['일시']
```

```python
def year(string):
  return string[:4]
def month(string):
  return string[5:7]
def day(string):
  return string[8:]


df['연'] = df['일시'].apply(year)
df['월'] = df['일시'].apply(month)
df['일'] = df['일시'].apply(day)
```

```python
df = df.drop(columns = ['일시'])
df
```

```python
import plotly.express as px
```

```python
df2 = df[df['연'].astype(int)>1999]
```

```python
df2
```

```python
df2['월일'] = df2['월'] + df2['일']
```

```python
df2 = df2[df2['월일'] != '0229']
```

```python
df2['x'] = df2.groupby('연').cumcount()+1
```

```python
df2
```

```python
px_plot = px.line(df2,x = 'x',  y = '평균기온', color = '연')
```

```python
px_plot
```

```python
col_mean = df.mean()
col_mean
```

```python
df_2 = df.fillna(col_mean)
```

```python
df_2.isna().sum()
```

```python
y = df_2['평균기온']
x = df_2.drop(columns = ['평균기온'])
```

```python
x
```

```python
y
```

```python
x['일시'].astype(str)
```

```python
x['일시']
def month(string):
  return int(string[-5:-3])
def day(string):
  return int(string[-2:])
x['월'] = x["일시"].apply(month)
x['일'] = x["일시"].apply(day)
```

```python

x
```

```python
from tensorflow import keras
```

```python
x.shape
```

```python
y.shape
```

```python
x_array = np.array(x)
x_array.shape
```

```python
type(x_array)
```

```python
y_array = np.array(y)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_array, y_array, test_size = 0.2)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
model = Sequential()
model.add(Dense(10, input_shape = (11,), activation = 'relu', name = 'input'))
model.add(Dense(10,  activation = 'relu', name = 'hidden'))
model.add(Dense(1, name = 'output'))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_absolute_error', metrics = ['mae'])
```

```python
model.summary()
```

```python
model.fit(X_train, Y_train, batch_size = 100, epochs = 100, verbose = 1)
```

```python
results = model.evaluate(X_test, Y_test)
results
```

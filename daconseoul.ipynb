```python
import kerastuner as kt
```

```python
def modelbuilder(hp):
  model = keras.Sequential()
  hp_units = hp.Int('units', min_value = 10, max_value = 200, step = 10)
  model.add(Dense(units = hp_units, input_shape = (11,), activation = 'relu', name = 'input'))
  model.add(Dense(units = hp_units, activation = 'relu', name = 'hidden'))
  model.add(Dense(1, name = 'output'))
  hp_learning_rate = hp.Choice('learning_rate', values = [0.1, 0.01, 0.001])
  model.compile(optimizer = keras.optimizers.Adam(learning_rate = hp_learning_rate), loss = 'mean_absolute_error', metrics = ['mae'])
  return model
```

```python
tuner = kt.Hyperband(modelbuilder,
                     objective = 'mae',
                     max_epochs = 10,
                     directory = 'my_dir',
                     project_name='hyperband')
```

```python
import IPython
```

```python
class cleartrainoutput(keras.callbacks.Callback):
  def on_train_end(*args, **kwargs):
    IPython.display.clear_output(wait = True)
```

```python
tuner.search(X_train, Y_train,
             epochs = 50,
             validation_data = (X_test, Y_test),
             callbacks = [cleartrainoutput()])

best_hps = tuner.get_best_hyperparameters(num_trials = 10)[0]
units, learning_rate = best_hps.get('units'), best_hps.get('learning_rate')
print(units, learning_rate)
```

```python
new_model = tuner.hypermodel.build(best_hps)
new_model.summary()
```

```python
new_model.fit(X_train, Y_train, epochs = 50, validation_data = (X_test, Y_test))
```

```python
temp4 = new_model.predict(X_test)
temp4 = temp4[-366:-8]
submission = pd.read_csv('./sample_submission.csv')
submission['평균기온'] = temp4
submission.to_csv('submission_4.csv', index=False)
```
